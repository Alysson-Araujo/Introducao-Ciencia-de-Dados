```
avaliacoes = pd.read_csv("https://github.com/Alysson-UFC/Introducao-Ciencia-de-Dados/blob/master/QuarentenaDados/Aula0/ml-latest-small/ratings.csv?raw=true") 
```

```
avaliacoes.head()
```

```
# saber o formato do data frame:
avaliacoes.shape
# Como pode ser visto no resultado, esse DataFrame possui 100836 linhas e 4 colunas.
```

```
# Forma de renomear o nome das colunas:
avaliacoes.columns = ["usuarioId", "filmeId", "nota", "momento"]
```

```
avaliacoes.head()
```

```
#dá uma breve descrição sobre o que tem dentro do DataFrame.
avaliacoes.describe()
```

```
#Pega uma determinada coluna e exibe seus dados em forma de série.
avaliacoes ["nota"]
```

```
avaliacoes.query
```

```
# busca as notas, de um determinado filme, de todos os usuários que deram alguma nota para ele
avaliacoes.query("filmeId==1")
```

```
# Busca os dados referente ao filme passado no parâmentro do query usando o .describe()
avaliacoes.query("filmeId==1").describe()
```

```
# Busca todos os dados do filme que possui Id==1, separa apenas a coluna que remete as notas
# desse filme e em seguida faz uma média usando o método .mean()
avaliacoes.query("filmeId==1")["nota"].mean()
```

Nesse momento iremos fazer um exemplo de como criar uma tabela com os nomes dos filmes e suas notas médias.

```
notas_medias_por_filme = avaliacoes.groupby("filmeId")["nota"].mean()
notas_medias_por_filme.head()
```
